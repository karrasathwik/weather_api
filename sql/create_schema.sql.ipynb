CREATING THE CATALOG


```sql
CREATE CATALOG IF NOT EXISTS cricket_lakehouse;
```

CREATING THE SCHEMA

```sql
CREATE SCHEMA IF NOT EXISTS cricket_lakehouse.config;
CREATE SCHEMA IF NOT EXISTS cricket_lakehouse.raw_layer;
CREATE SCHEMA IF NOT EXISTS cricket_lakehouse.bronze_layer;
CREATE SCHEMA IF NOT EXISTS cricket_lakehouse.silver_layer;
CREATE SCHEMA IF NOT EXISTS cricket_lakehouse.gold_layer;
-- SCHEMAS CREATED: config, raw_layer, bronze_layer, silver_layer, gold_layer
```
