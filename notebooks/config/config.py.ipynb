```python
def load_pipeline_config(pipeline_config,config_table):
    print("=" * 50)
    print("loading api_config")
    print(f"table:{config_table}")
    print(f"pipelines : {len(pipeline_config)}")
    df = spark.createDataFrame(pipeline_config)
    df = df.withColumn("created_date",current_timestamp())
    df.write \
        .format("delta") \
        .mode("overwrite") \
        .option("overwriteSchema", "true") \
        .saveAsTable(config_table)
    count = spark.table(config_table).count()
    print(f"\n✅ {count} pipelines loaded!")

    print("\nMetadata Table:")
    spark.table(config_table).show(truncate=False)


load_pipeline_config(
    pipeline_config = pipeline_config,
    config_table     = config_table
)
```
